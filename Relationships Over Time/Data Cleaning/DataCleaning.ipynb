```
import pandas as pd

status = pd.read_csv(r'status.csv')
marry = pd.read_csv(r'marry.csv')
# df.to_csv('trial.csv', index=False)
```

```
marry['3'] = marry['3'].str.lower()
marry['4'] = marry['4'].str.lower()
marry = marry.replace('not married', 'partnered')
marry = marry.replace('unmarried partners', 'partnered')

status = status.replace('still together at w5 or some follow-up w/o breakup', 'still together')
status = status.replace('broke up at wave 2,3,4, or 5', 'broke up')
status = status.replace('partner deceased', 'partner passed away')

marry = marry[status['combo'].notna()]
status = status[status['combo'].notna()]
```

```
marry.to_csv('marry_new.csv', index=False)
status.to_csv('status_new.csv', index=False)
```

```
codes = []

for i in range(len(marry)):
    m = list(marry.iloc[i])[1:]
    s = list(status.iloc[i])
    code = []

    did_marry = 1
    if 'married' in m:
        did_marry = 10

    if 'partner passed away' in s:
        code.append('False')
        code.append(-1 * did_marry *(s.index('partner passed away') + 1))

    elif s[-1] == 'still together':
        code.append('True')
        if m[0] == 'married' and m[4] == 'married':
            code.append(0)
        elif m[0] == 'partnered' and m[4] == 'partnered':
            code.append(-1)
        else:
            for j in range(4):
                if m[j] == 'partnered' and m[j+1] == 'married':
                    code.append(j+1)  
    else:
        code.append('False')
        code.append(did_marry * (s.index('broke up') + 1))


    if len(code) == 2:
        codes.append(code)
    

```

```
cols = ['st', 'type']
import csv
with open('codes.csv', 'w') as f:
    write = csv.writer(f)
    write.writerow(cols)
    write.writerows(codes)
```
